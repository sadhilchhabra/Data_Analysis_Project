# Data Visualization

### Gender(sex)

```python
# Calculated counts
sex_counts = df['sex'].value_counts()

# Custom function to display the number and percentage
def autopct_func(pct):
    total = sum(sex_counts)
    val = int(round(pct * total / 100.0))
    return f'{val} ({pct:.2f}%)'

# Pie Chart
colors = cm.get_cmap('tab20c')(np.linspace(0, 1, len(sex_counts)))
plt.figure(figsize=(8, 8))
plt.title('Gender (Sex) Distribution', fontsize=16)
plt.pie(sex_counts, labels=sex_counts.index, autopct=autopct_func, wedgeprops=dict(alpha=0.8), colors=colors)
plt.show()
```

### Heart Disease Based on Gender

```python
# Define the labels for x-axis and legend
gender_labels = {0: 'Male', 1: 'Female'}
target_labels = ['False', 'True']

# Bar Chart
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='sex', hue='target', data=df, palette='cool')
plt.title('Heart Disease Distribution Based on Gender', fontweight='bold', fontsize=16)
plt.xlabel('Gender (Sex)', fontweight='bold')
plt.ylabel('Total', fontweight='bold')

# Annotate bars with counts
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')

# Update x-axis labels
ax.set_xticklabels([gender_labels[int(label.get_text())] for label in ax.get_xticklabels()])

# Customize legend
plt.legend(title='Heart Disease', title_fontsize='13', fontsize='10', loc='upper left')

# Grid settings
plt.grid(axis='y', alpha=0.4)
plt.tight_layout()
plt.show()
```

### Heart Disease Based on Fasting Blood Sugar

```python
# Labels for the x-axis and legend
fbs_labels = ['< 120 mg/dl', '> 120 mg/dl']
target_labels = ['No Heart Disease', 'Heart Disease']

# Bar Chart
plt.figure(figsize=(12, 7))  
ax = sns.countplot(x='fbs', hue='target', data=df, palette='viridis')

# Annotate bars with counts
for rect in ax.patches:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 1, f'{int(height)}', 
            ha='center', va='bottom', fontsize=10)

# MetaData
plt.suptitle('Heart Disease Distribution Based on Fasting Blood Sugar', fontweight='bold', fontsize=16)
plt.xlabel('Fasting Blood Sugar', fontweight='bold')
plt.ylabel('Total', fontweight='bold')
plt.xticks(ticks=range(len(fbs_labels)), labels=fbs_labels, rotation=0)
plt.grid(axis='y', alpha=0.4)
plt.legend(title='Heart Disease Status', title_fontsize='11', fontsize='10', loc='upper left', bbox_to_anchor=(1.05, 1))
plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```
